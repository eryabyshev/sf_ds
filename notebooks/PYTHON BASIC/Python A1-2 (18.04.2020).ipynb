## Цикл for . Цикл с параметром##

Существует три типичных сценария использования for.

Организация цикла

```python
#Перебор числе в заданном диапазоне и вывод на экран текущего значения параметра цикла 
n = int(input("Введите число"))
for i in range(n):
    print(i)
```

## Итерируемые объекты

Цикл принимает список чисел и перебирает его по одному элементу, выводя текущее значение на экран.

```python
#Перебор элементов списка
a = [1,2,3,4,5]
for i in range(len(a)):
    print(i)
```

Цикл for проходит по строке, каждый символ которой обрабатывается в отдельной итерации.

```python
#Перебор элементов строки
string = 'SkillFactory'
for i in string:
    print(i, end = '*')
```

```python
# Использование enumerate
#Позиция и значение

string = 'Программирование'
```

```python
for x in enumerate(string):
    print(x)
```

```python
for x, i in enumerate(string):
    print(x, i)
```

Давайте решим задачку!
Есть программа, которая проверяет, находится ли текущее значение минут в заданном списке.
Нужно выполнить проверку 5 раз

```python
from datetime import datetime

date = [x for x in range(1,60,2)]

right_minute = datetime.today().minute
for i in range(5):
    if right_minute in date:
        print("Yes")
    else:
        print('No')
```

Гипотеза Сиракуз гласит, что любое натуральное число сводится к единице при следующих действиях над ним: 

а) если число четное, то разделить его пополам, 

б) если число нечетное, то умножить его на 3, прибавить 1 и результат разделить на 2. 

Над вновь полученным числом вновь повторить действия a) или б) в зависимости от четности числа. Рано или поздно число станет равным 1.

Напишем программу, проверяющую гипотезу Сиракуз. Пусть пользователь вводит любое натуральное число. Будем выполнять в цикле над ним вышеуказанные действия, пока оно не станет равным 1. Если гипотеза верна, то рано или поздно цикл закончится, а если нет - то произойдет зацикливание программы.

## Цикл while ##

Цикл с предусловием


«Цикл while» - это цикл с условием, которые повторяет блок команд до тех пор, пока условие не будет выполнено.

Цикл while может использоваться для перебора числовой последовательности.

Например, следующий цикл считает от 1 до 5:

```python
current_number = 1

while current_number <= 5: 
    print(current_number)    
    current_number += 1
```

```python
a = True
i=1
while a == True:
    print(i)
    i+=1
    if i == 5:
        a = False
```

## Задание "Числа Фибоначчи"##

Последовательность Фибоначчи определяется так:
φ_0 = 1,  φ_1 = 1,  φ_n = φ_n−1 + φ_n−2.

По данному числу n определите n-е число Фибоначчи φn.

```python
n=int(input('введите число: '))
Fn_1=1
Fn_2=1
Fn=0
for i in range(2,n+1):
    Fn=Fn_2+Fn_1
    #print('Fn=',Fn)
    Fn_1=Fn_2
    #print('Fn_1=',Fn_1)
    Fn_2=Fn
    #print('Fn_2=',Fn_2)
else:
    print(Fn)
#1 1 2 3 5 8
```

## Инструкции управления циклом ##
После тела цикла можно написать слово else: и после него блок операций, который будет выполнен один раз после окончания цикла, когда проверяемое условие станет неверно

```python
i = 1
while i <= 10:
    print(i)
    i+=1
else:
    print("Цикл окончен, i =", i)
```

```python
for i in 'hello world':
    if i == 'a':
        break
else:
    print('Буквы a в строке нет')
```
